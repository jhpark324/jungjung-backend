```python
import os
from dotenv import load_dotenv
load_dotenv()

upstage_api_key = os.getenv("UPSTAGE_API_KEY")
```

```python
from langchain_upstage import UpstageDocumentParseLoader

file_path = "../backend/documents/국회본회의_회의록._제429회(14차).pdf"
layzer = UpstageDocumentParseLoader(file_path, output_format="text")

docs = layzer.load()

```

```python
with open("output.txt", "w", encoding="utf-8") as f:
    for doc in docs:
        f.write(doc.page_content)
        f.write("\n\n---\n\n")  
```
